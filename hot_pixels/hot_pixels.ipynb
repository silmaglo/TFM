We will add hot pixels (N pixels with a lot of energy in horizontal or vertical) to the random energy/noise images.

```python
window_x, window_y = all_electron_data.shape[2], all_electron_data.shape[1]
```

```python
#'index' is the number of the file, and 'file' is the name of the file
for index, file in enumerate(all_electron_data): 
    a = np.random.rand()
    
    if (a >= 0.5): #vertical line
        
        if (a>=0.75): #horizontal line
            print('horizontal line')
            energy_a = file
            
            #how long will the line be
            long = np.random.randint(0, window_x) 
            
            #random coordinates
            y_1 = np.random.randint(0, window_y)
            x_1 = np.random.randint(0, window_x-long)
            
            #filling pixels
            energy_b = np.copy(energy_a)
            
            for x in range(x_1, x_1+long+1):   
                energy_b[y_1, x] = 18000
            
            np.savez('file_index' + str(index), energy=energy_b)
            #np.savez(os.path.join(path, files[index]), energy=energy_b)
            
        else:
            print('vertical line')
            energy_a = file
            long = np.random.randint(0,window_y)
            
            #random coordinates
            y_1 = np.random.randint(0, window_y-long)
            x_1 = np.random.randint(0, window_x)
            
            #filling pixels
            energy_b = np.copy(energy_a)
            
            for y in range(y_1, y_1+long+1):
                energy_b[y, x_1] = 18000
            
            np.savez('file_index' + str(index), energy=energy_b)
            #np.savez(os.path.join(path, files[index]), energy=energy_b)
    else:
        print('no line')
        energy_a = file
        np.savez('file_index' + str(index), energy=energy_a)
        #np.savez(os.path.join(path, files[index]), energy=energy_a)
        #np.savez('file_index%d' + str(index), energy=energy_a)
```
